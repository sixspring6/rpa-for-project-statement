```python
#네이버 검색

from selenium import webdriver
from selenium.webdriver.edge.service import Service
from webdriver_manager.microsoft import EdgeChromiumDriverManager
from selenium.webdriver.common.by import By

# Edge WebDriver 초기화
driver = webdriver.Edge(service=Service(EdgeChromiumDriverManager().install()))

# 원하는 URL로 이동
url = "https://www.naver.com"  # 여기에 접속하고 싶은 URL을 입력하세요.
driver.maximize_window()
driver.get(url)

query = "python"
input_element = driver.find_element(By.NAME, "query") # 검색어를 입력할 수 있는 검색창 찾기
input_element.clear()      
input_element.send_keys(query)                    # 검색창에 검색어 입력 
input_element.submit()   

# 이후 필요한 작업 수행 (예: 웹 페이지 내용 검색, 폼 제출 등)

# 작업 완료 후 브라우저 닫기
print("- 접속한 웹 사이트의 제목:", driver.title) # 접속한 웹 사이트의 제목 출력
print("- 접속한 웹 사이트의 URL:", driver.current_url) # 접속한 웹 사이트의 URL 출력
```
